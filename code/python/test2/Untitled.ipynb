```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,200,200)
V = np.zeros(200)
P = np.zeros(200)
# for i in range(0,5): # i: 0,1,2,3,4

tau = 60
V[0] = 200
n = len(V)


for i in range(0,n-1):
    V[i+1] = V[i] + tau*0.001*(400-V[i]);


P[0] = 400
for i in range(0,n-1):
    P[i+1] = P[i] + 10*0.01*(350-1*P[i]);

    
plt.plot(t, V ,label='V')
plt.grid()
plt.show()

plt.plot(P)
plt.grid()

```
